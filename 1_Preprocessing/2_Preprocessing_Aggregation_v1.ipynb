# Import Libraries & Data

```python
import os
import glob
import re
import numpy as np
import pandas as pd
```

```python
dataset_au = pd.read_pickle('dolos_openface_merged.pkl')
```

# Aggregate all data again

```python
# Here we use only the AU columns

au_cols = [c for c in dataset_au.columns if c.endswith("_r")]
```

```python
# compute aggregates 

# compute AU mean per video
video_mean = (dataset_au.groupby('video_id')[au_cols].mean())
video_mean

# compute AU max per video
video_max = (dataset_au.groupby('video_id')[au_cols].max())
video_max

# compute AU std per video
video_std = (dataset_au.groupby('video_id')[au_cols].std())
video_std
```

```python
# bring label column as well

labels = (dataset_au[['video_id', 'label']].drop_duplicates(subset = 'video_id'))

dataset_mean = video_mean.merge(labels, on = 'video_id')
dataset_mean

dataset_max = video_max.merge(labels, on = 'video_id')
dataset_max

dataset_std = video_std.merge(labels, on = 'video_id')
dataset_std
```

#### ---> From now on we have one row per video with AU mean/max/std & label.

# Output Files

```python
dataset_mean.to_pickle('dolos_aggr_mean.pkl')
dataset_max.to_pickle('dolos_aggr_max.pkl')
dataset_std.to_pickle('dolos_aggr_std.pkl')
```
