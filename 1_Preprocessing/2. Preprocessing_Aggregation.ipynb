# Import Libraries & Data

```python
import os
import glob
import re
import numpy as np
import pandas as pd
```

```python
dataset_au = pd.read_pickle('dolos_openface_merged.pkl')
```

# Aggregate all data again

```python
dataset_au['video_id'] = dataset_au['file_name'].str.replace('.csv', '', regex = False)
```

```python
# keep only au columns

au_r_cols = [c for c in dataset_au.columns if c.endswith("_r")]
au_r_cols
```

```python
# compute AU mean per video

video_means = (dataset_au.groupby('video_id')[au_r_cols].mean())
video_means
```

```python
# bring label column as well

labels = (dataset_au[['video_id', 'label']].drop_duplicates(subset = 'video_id'))
dataset_static = video_means.merge(labels, on = 'video_id')
dataset_static
```

# Output Files

```python
dataset_static.to_pickle('dolos_aggr_mean.pkl')
```
