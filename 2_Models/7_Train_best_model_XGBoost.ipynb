# 1. Import Libraries and Data

```python
import joblib
import pandas as pd
from xgboost import XGBClassifier
```

# 2. Train Best model - XGBoost

```python
# previous results - params: {'n_estimators': 500, 'max_depth': 6, 'learning_rate': 0.01, 'subsample': 0.8, 'colsample_bytree': 0.7}, CV AUC: 0.6449
```

```python
best_params = {
    "n_estimators": 500,
    "max_depth": 6,
    "learning_rate": 0.01,
    "subsample": 0.8,
    "colsample_bytree": 0.7
}

# read all training data
df = pd.read_pickle('/home/maria/Desktop/Deception_project/00_Datasets_to_run/dolos_aggr_mean_v2.pkl')

# prepare features and labels
X = df.drop(columns=['label', 'face_id', 'video_id'])
X = X.select_dtypes(include=['number']).to_numpy().astype('float32')
y = df['label'].map({'lie': 0, 'truth': 1}).values

# train classifier on ALL data
clf = XGBClassifier(
    objective="binary:logistic",
    eval_metric="auc",
    tree_method="hist",
    random_state=0,
    n_jobs=-1,
    **best_params
)

clf.fit(X, y)

# save model
joblib.dump(clf, "xgb_final_model.pkl")
print("Model saved as xgb_final_model.pkl")
```
